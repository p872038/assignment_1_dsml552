```python
from sklearn.metrics import accuracy_score
import math
x = []
y = []
min_test_error = []
N = range(10, 211,10)
all_one = df.loc[(df['class'] == 1)]
all_zero = df.loc[(df['class']== 0)]
for i in N:
    count_zero = round(i/3)
    count_one = (i - count_zero)
    train_one = all_one[:int(count_one)]
    train_zero = all_zero[:int(count_zero)]
    test_one = all_one[int(count_one):]
    test_zero = all_zero[int(count_zero):]
    X_train = pd.concat([train_one, train_zero]).drop(labels='class', axis=1)
    Y_train = pd.concat([train_one, train_zero])["class"].values
    X_test =  pd.concat([test_one, test_zero]).drop(labels='class', axis=1)
    Y_test = pd.concat([test_one, test_zero])["class"].values
    k = 1
    bestscore = 0
    bestk = 1
    while k<i:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train,Y_train)
        pred_i = knn.predict(X_test)
        #print(accuracy_score(Y_test,pred_i))
        if(bestscore < accuracy_score(Y_test,pred_i)*100):
            bestscore = accuracy_score(Y_test,pred_i)*100
            bestk = k
        k = k + 5
     
    print("for the given N =  ", i , "the best score is ", bestscore, "for the given k ", bestk, "\n")
    y.append(i)
    x.append(bestk)
    min_test_error.append((100-bestscore)/100)
    

# 170 is the value of N 
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

#plt.scatter(y,x, color = "blue")
#plt.show()
# plot learning curve

plt.figure(figsize = (20,8))
plt.plot(range(10,211,10),min_test_error,color='green', 
         marker='o',markerfacecolor='blue', markersize=6)
plt.title('LEARNING CURVE')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
```

```python
# Mahalanobis distance

from sklearn.metrics import mean_absolute_error
import math
min_test_error = []
bestScore = 0
min_error = 100
bestk = 1
for k in range(1,197,5) : 
    knn =  KNeighborsClassifier(n_neighbors = k, metric = "mahalanobis", metric_params = {'VI' : np.linalg.inv(np.cov(X_train,rowvar=False))})
    knn.fit(X_train, Y_train)
    pred_k = knn.predict(X_test)
    min_test_error.append(np.mean(pred_k != Y_test))
    if(min_error > (mean_absolute_error(Y_test, pred_k))):
        min_error = mean_absolute_error(Y_test, pred_k)
        bestk = k


print("the best k is ", bestk, "with the error_rate",min_error)
plt.figure(figsize = (20,8))
plt.plot(range(1,197,5),min_test_error,color='green', 
         marker='o',markerfacecolor='blue', markersize=6)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
```
